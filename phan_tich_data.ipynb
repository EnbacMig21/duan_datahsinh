### 1.1 Phân tính kết quả theo kết quả học tập.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu từ file của bạn
df = pd.read_csv('student_info_updated.csv')

# 2. Định nghĩa từ điển quy đổi (Mapping) chuẩn xác theo ảnh bạn gửi
mapping_edu = {
    1: 'High School',
    2: "Bachelor's",
    3: "Master's",
    4: 'PhD',
    '1': 'High School',
    '2': "Bachelor's",
    '3': "Master's",
    '4': 'PhD'
}

# Tạo cột nhãn chữ mới
df['parent_education_label'] = df['parent_education'].map(mapping_edu)
df['parent_education_label'] = df['parent_education_label'].fillna(df['parent_education'])

# 3. Tính điểm trung bình của từng môn theo nhãn chữ mới
subject_scores_by_parent = df.groupby('parent_education_label')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()
subject_scores_by_parent = subject_scores_by_parent.round(2)

# Sắp xếp thứ tự hiển thị từ thấp đến cao theo đúng logic học thuật
ordered_categories = ['High School', "Bachelor's", "Master's", 'PhD']
subject_scores_by_parent['parent_education_label'] = pd.Categorical(
    subject_scores_by_parent['parent_education_label'], 
    categories=ordered_categories, 
    ordered=True
)
subject_scores_by_parent = subject_scores_by_parent.sort_values('parent_education_label').reset_index(drop=True)

# Đổi tên các cột sang tiếng Việt để bảng in ra sạch sẽ
subject_scores_by_parent.columns = ['Cấp học phụ huynh', 'Điểm Toán TB', 'Điểm Đọc TB', 'Điểm Viết TB']

# 4. In kết quả số cụ thể ra màn hình
print("--- KẾT QUẢ SỐ CỤ THỂ: ĐIỂM 3 MÔN THEO CẤP HỌC PHỤ HUYNH ---")
print(subject_scores_by_parent.to_string(index=False))

# ==========================================================
# 5. TRỰC QUAN HÓA (Vẽ biểu đồ cột nhóm)
# ==========================================================
df_melted = pd.melt(subject_scores_by_parent, 
                    id_vars=['Cấp học phụ huynh'], 
                    value_vars=['Điểm Toán TB', 'Điểm Đọc TB', 'Điểm Viết TB'],
                    var_name='Môn học', 
                    value_name='Điểm số trung bình')

plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Vẽ biểu đồ cột nhóm
ax = sns.barplot(x='Cấp học phụ huynh', y='Điểm số trung bình', hue='Môn học', data=df_melted, palette='Set2')

# Cấu hình tiêu đề và nhãn trục
plt.title('THỐNG KÊ CHI TIẾT ĐIỂM 3 MÔN THEO CẤP HỌC PHỤ HUYNH', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Trình độ học vấn của phụ huynh', fontsize=11, labelpad=10)
plt.ylabel('Điểm trung bình (Thang điểm 100)', fontsize=11)
plt.ylim(0, 115) 

# Hiển thị số điểm cụ thể lên đầu từng cột
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'{p.get_height():.1f}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', 
                    xytext=(0, 8), 
                    textcoords='offset points', 
                    fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

### 1.2 Phân tích theo hành vi học tập 

```python
import pandas as pd

# 1. Đọc file dữ liệu của bạn
df = pd.read_csv('Bao_Cao_Du_Doan_Hoc_Sinh.csv')

# Chuẩn hóa tên cột: bỏ khoảng trắng thừa và chuyển về chữ thường
df.columns = df.columns.str.strip().str.lower()

# Tạo thêm cột tỷ lệ chuyên cần theo % để dễ đọc số liệu
df['attendance_rate_pct'] = df['attendance_rate']

# Định nghĩa danh sách 7 nhân tố cần phân tích theo đúng tên cột thực tế
cac_nhan_to = {
    'extra_activities': '1. HOẠT ĐỘNG NGOẠI KHÓA',
    'lunch_type': '2. CHẾ ĐỘ ĂN TRƯA',
    'internet_access': '3. TRUY CẬP INTERNET TẠI NHÀ',
    'gender': '4. GIỚI TÍNH',
    'grade_level': '5. KHỐI LỚP',
    'age': '6. TUỔI',
    'parent_education': '7. HỌC VẤN PHU HUYNH'
}

print("=========================================================================")
print("  BÁO CÁO SỐ LIỆU CHỈ SỐ CHĂM CHỈ & CHUYÊN CẦN THEO 7 YẾU TỐ TÁC ĐỘNG  ")
print("=========================================================================\n")

# Vòng lặp duyệt qua từng nhân tố để tính toán và xuất bảng số liệu
for col_name, ten_tieng_viet in cac_nhan_to.items():
    if col_name in df.columns:
        # Gom nhóm và tính toán giá trị trung bình đồng thời cho cả 2 chỉ số
        df_summary = df.groupby(col_name)[['study_hours', 'attendance_rate_pct']].mean()
        
        # Đổi tên cột sang Tiếng Việt cho sạch đẹp để bạn dễ copy vào slide
        df_summary.columns = ['Số giờ tự học TB (Giờ/Tuần)', 'Tỷ lệ chuyên cần TB (%)']
        
        # Làm tròn số liệu đến 2 chữ số thập phân
        df_summary = df_summary.round(2)
        
        # In kết quả ra màn hình dạng bảng
        print(f"--- BẢNG SỐ LIỆU THEO: {ten_tieng_viet} ---")
        print(df_summary)
        print("=" * 73 + "\n")
```

Chúng ta cũng thấy những học sinh có tham gia hoạt động ngoại khóa có số giờ học TB cao hơn nhóm còn lại, đồng thời cũng có nhiều học sinh chăm học hơn bên có tham gia hoạt động ngoại khóa.

## 2. Học sinh có yếu tố học lệch môn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc file báo cáo dữ liệu của bạn
df = pd.read_csv('Bao_Cao_Du_Doan_Hoc_Sinh.csv')
df.columns = df.columns.str.strip().str.lower()

# ==========================================================
# THỰC HIỆN PHÂN LOẠI TOÀN DIỆN THEO TIÊU CHÍ CỦA BẠN
# ==========================================================
def phan_loai_hoc_sinh(row):
    math = row['math_score']
    reading = row['reading_score']
    writing = row['writing_score']
    total = math + reading + writing
    
    if total == 0:
        return 'Chưa có điểm'
        
    # Tiêu chí 1: Học tốt cả 3 môn (> 80)
    if math > 80 and reading > 80 and writing > 80:
        return 'Học tốt cả 3 môn'
        
    # Tiêu chí 2: Học kém cả 3 môn (< 70)
    if math < 70 and reading < 70 and writing < 70:
        return 'Học kém cả 3 môn'
        
    # Tiêu chí 3: Kiểm tra điều kiện học đều (Hệ số 0.265 của bạn)
    p_math = math / total
    p_reading = reading / total
    p_writing = writing / total
    
    # Nếu không có môn nào < 26.5% tổng điểm -> Học đều
    if p_math >= 0.265 and p_reading >= 0.265 and p_writing >= 0.265:
        return 'Học đều 3 môn'
        
    # Tiêu chí 4: Phần còn lại (Học lệch)
    return 'Phần còn lại (Học lệch)'

# Áp dụng hàm phân loại
df['Phân loại năng lực'] = df.apply(phan_loai_hoc_sinh, axis=1)

# Thống kê số lượng học sinh ở từng nhóm
df_count = df['Phân loại năng lực'].value_counts().reset_index()
df_count.columns = ['Nhóm năng lực', 'Số lượng học sinh']

# Sắp xếp thứ tự hiển thị trên biểu đồ cho logic và đẹp mắt
order_list = ['Học tốt cả 3 môn', 'Học đều 3 môn', 'Phọc kém cả 3 môn', 'Phần còn lại (Học lệch)']
# Sửa lại chính xác tên để khớp bảng dữ liệu
order_list = ['Học tốt cả 3 môn', 'Học đều 3 môn', 'Học kém cả 3 môn', 'Phần còn lại (Học lệch)']

print("--- THỐNG KÊ SỐ LƯỢNG HỌC SINH THEO PHÂN LOẠI MỚI ---")
print(df_count.to_string(index=False))

# ==========================================================
# 2. TRỰC QUAN HÓA BẰNG BIỂU ĐỒ CỘT (BAR CHART)
# ==========================================================
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Định nghĩa bảng màu sang trọng cho slide: Xanh lá, Xanh dương, Xám, Đỏ cam
custom_colors = ['#2ecc71', '#3498db', '#95a5a6', '#e67e22']

ax = sns.barplot(x='Nhóm năng lực', y='Số lượng học sinh', data=df_count, 
                 order=order_list, palette=custom_colors, width=0.5)

# Cấu hình tiêu đề và nhãn các trục
plt.title('PHÂN PHỐI SỐ LƯỢNG HỌC SINH THEO CÁC NHÓM NĂNG LỰC HỌC TẬP', fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Nhóm phân loại học sinh', fontsize=11, labelpad=12)
plt.ylabel('Số lượng học sinh (HS)', fontsize=11, labelpad=12)

# --- THỦ THUẬT CĂN CHỈNH BỐ CỤC LỀ CÂN ĐỐI ---
plt.margins(x=0.1) # Tạo khoảng trống vừa phải với lề biên
plt.ylim(0, df_count['Số lượng học sinh'].max() * 1.15) # Tạo khoảng trống phía trên để ghi số không bị lẹm khung

# Ghi số lượng học sinh cụ thể lên đầu từng cột
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'{int(p.get_height())} HS', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='bottom', 
                    xytext=(0, 5), textcoords='offset points', 
                    fontsize=11, fontweight='bold', color='#2c3e50')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc file báo cáo dữ liệu của bạn
df = pd.read_csv('Bao_Cao_Du_Doan_Hoc_Sinh.csv')
df.columns = df.columns.str.strip().str.lower()

# ==========================================================
# KHỞI TẠO LOGIC PHÂN LOẠI CHI TIẾT 6 NHÓM
# ==========================================================
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['math_low'] = df['math_score'] < (0.265 * df['total_score'])
df['reading_low'] = df['reading_score'] < (0.265 * df['total_score'])
df['writing_low'] = df['writing_score'] < (0.265 * df['total_score'])

def phan_loai_chi_tiet_6_nhom(row):
    math = row['math_score']
    reading = row['reading_score']
    writing = row['writing_score']
    
    if math > 80 and reading > 80 and writing > 80: return 'Học tốt cả 3 môn'
    if math < 70 and reading < 70 and writing < 70: return 'Học kém cả 3 môn'
    if row['math_low']: return 'Học lệch - Yếu Toán'
    if row['reading_low']: return 'Học lệch - Yếu Đọc'
    if row['writing_low']: return 'Học lệch - Yếu Viết'
    return 'Học đều 3 môn'

df['Phân loại năng lực chi tiết'] = df.apply(phan_loai_chi_tiet_6_nhom, axis=1)

# ==========================================================
# 2. TÍNH ĐIỂM TRUNG BÌNH VÀ XOAY BẢNG (MELT) ĐỂ VẼ TRỤC NGANG
# ==========================================================
df_diem_tb = df.groupby('Phân loại năng lực chi tiết')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()

# Đổi cấu trúc bảng sang dạng Long-format để Seaborn hiểu trục Y và trục X
df_melted_tb = pd.melt(df_diem_tb, 
                       id_vars=['Phân loại năng lực chi tiết'], 
                       value_vars=['math_score', 'reading_score', 'writing_score'],
                       var_name='Môn học', value_name='Điểm trung bình')

df_melted_tb['Môn học'] = df_melted_tb['Môn học'].map({
    'math_score': 'Toán (Math)', 
    'reading_score': 'Đọc (Reading)', 
    'writing_score': 'Viết (Writing)'
})

# Sắp xếp thứ tự các hàng (nhóm) cố định trên trục Y
order_6_groups = [
    'Học tốt cả 3 môn', 'Học đều 3 môn', 'Học lệch - Yếu Toán',
    'Học lệch - Yếu Đọc', 'Học lệch - Yếu Viết', 'Học kém cả 3 môn'
]

# ==========================================================
# 3. TRỰC QUAN HÓA BIỂU ĐỒ CỘT NẰM NGANG
# ==========================================================
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")

# Bảng màu nhất quán: Toán (Cam), Đọc (Xanh dương), Viết (Tím)
colors = {'Toán (Math)': '#e67e22', 'Đọc (Reading)': '#3498db', 'Viết (Writing)': '#9b59b6'}

# Vẽ biểu đồ với x là điểm trung bình (cột) và y là nhóm học sinh (hàng)
ax = sns.barplot(data=df_melted_tb, y='Phân loại năng lực chi tiết', x='Điểm trung bình', hue='Môn học',
                 order=order_6_groups, palette=colors, edgecolor='white', errorbar=None, width=0.7)

# Cấu hình tiêu đề và nhãn trục
plt.title('BẢNG ĐIỂM TRUNG BÌNH CỦA CÁC MÔN CHIA THEO CÁC HÀNG NHÓM HỌC SINH', fontsize=14, fontweight='bold', pad=20)
plt.ylabel('Các nhóm học sinh (Hàng)', fontsize=11, labelpad=12)
plt.xlabel('Điểm số trung bình thực tế (Cột)', fontsize=11, labelpad=12)

# Giới hạn trục X để thấy rõ độ chênh lệch (từ 40 đến 100 điểm)
plt.xlim(40, 100)
plt.legend(title='Môn học', loc='lower right', frameon=True)

# Hiển thị số điểm chính xác ngay bên cạnh mỗi cột nằm ngang
for p in ax.patches:
    w = p.get_width()
    if w > 0:
        ax.annotate(f'{w:.1f}', 
                    (w, p.get_y() + p.get_height() / 2.), 
                    ha='left', va='center', 
                    xytext=(5, 0), textcoords='offset points', 
                    fontsize=9, fontweight='bold', color='#2c3e50')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc file báo cáo dữ liệu của bạn
df = pd.read_csv('Bao_Cao_Du_Doan_Hoc_Sinh.csv')
df.columns = df.columns.str.strip().str.lower()

# ==========================================================
# LỌC ĐÚNG NHÓM HỌC ĐỀU (Theo logic hệ số 0.265 của bạn)
# ==========================================================
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['p_math'] = df['math_score'] / df['total_score']
df['p_reading'] = df['reading_score'] / df['total_score']
df['p_writing'] = df['writing_score'] / df['total_score']

def xac_dinh_hoc_deu(row):
    math, reading, writing = row['math_score'], row['reading_score'], row['writing_score']
    # Loại bỏ nhóm tốt cả 3 (>80) và kém cả 3 (<70)
    if math > 80 and reading > 80 and writing > 80: return False
    if math < 70 and reading < 70 and writing < 70: return False
    # Kiểm tra điều kiện học đều
    if row['p_math'] >= 0.265 and row['p_reading'] >= 0.265 and row['p_writing'] >= 0.265:
        return True
    return False

# Lọc riêng dữ liệu nhóm học đều
df_hoc_deu = df[df.apply(xac_dinh_hoc_deu, axis=1)].copy()

# ==========================================================
# ĐẾM SỐ MÔN ĐẠT ĐIỂM TRÊN 80 CỦA TỪNG HỌC SINH
# ==========================================================
df_hoc_deu['math_gt_80'] = df_hoc_deu['math_score'] > 80
df_hoc_deu['reading_gt_80'] = df_hoc_deu['reading_score'] > 80
df_hoc_deu['writing_gt_80'] = df_hoc_deu['writing_score'] > 80

# Tổng số môn > 80 của mỗi em (kết quả chỉ có thể là 0, 1, hoặc 2 môn)
# (Vì kịch bản bằng 3 môn đã bị loại ngay từ đầu ở bước lọc nhóm học tốt)
df_hoc_deu['so_mon_tren_80'] = df_hoc_deu[['math_gt_80', 'reading_gt_80', 'writing_gt_80']].sum(axis=1)

# Định nghĩa nhãn hiển thị cho biểu đồ
df_hoc_deu['Phân loại chi tiết'] = df_hoc_deu['so_mon_tren_80'].map({
    0: 'Không môn nào trên 80',
    1: 'Có đúng 1 môn trên 80',
    2: 'Có đúng 2 môn trên 80'
})

# Thống kê số lượng cụ thể
df_result = df_hoc_deu['Phân loại chi tiết'].value_counts().reset_index()
df_result.columns = ['Tiêu chí phân loại', 'Số lượng học sinh']

# Sắp xếp theo thứ tự hiển thị logic
order_list = ['Không môn nào trên 80', 'Có đúng 1 môn trên 80', 'Có đúng 2 môn trên 80']
df_result = df_result.set_index('Tiêu chí phân loại').reindex(order_list).reset_index()

print("--- THỐNG KÊ CHI TIẾT BÊN TRONG NHÓM HỌC ĐỀU ---")
print(df_result.to_string(index=False))

# ==========================================================
# 3. TRỰC QUAN HÓA BẰNG BIỂU ĐỒ CỘT DỌC
# ==========================================================
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Bảng màu chuyển tiếp: Xám (Không môn), Xanh nhạt (1 môn), Xanh đậm (2 môn)
custom_colors = ['#95a5a6', '#5dade2', '#2e86c1']

ax = sns.barplot(data=df_result, x='Tiêu chí phân loại', y='Số lượng học sinh', 
                 palette=custom_colors, width=0.4, edgecolor='white')

# Cấu hình tiêu đề và nhãn
plt.title('CẤU TRÚC PHÂN RÃ CHI TIẾT CỦA NHÓM HỌC SINH HỌC ĐỀU 3 MÔN', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Số lượng môn học đạt điểm giỏi (> 80)', fontsize=11, labelpad=10)
plt.ylabel('Số lượng học sinh (Học sinh)', fontsize=11, labelpad=10)

# Căn chỉnh giới hạn trục Y tạo không gian ghi số liệu
plt.ylim(0, df_result['Số lượng học sinh'].max() * 1.15)

# Ghi số lượng học sinh cụ thể lên đầu từng cột
for p in ax.patches:
    h = p.get_height()
    if h > 0:
        ax.annotate(f'{int(h)} HS', 
                    (p.get_x() + p.get_width() / 2., h), 
                    ha='center', va='bottom', 
                    xytext=(0, 4), textcoords='offset points', 
                    fontsize=10, fontweight='bold', color='#2c3e50')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 1. Đọc file báo cáo dữ liệu của bạn
df = pd.read_csv('Bao_Cao_Du_Doan_Hoc_Sinh.csv')
df.columns = df.columns.str.strip().str.lower()

# ==========================================================
# 2. LỌC ĐIỀU KIỆN: ÍT NHẤT 1 MÔN > 90
# ==========================================================
# Điều kiện kiểm tra từng môn độc lập
cond_math = df['math_score'] > 90
cond_reading = df['reading_score'] > 90
cond_writing = df['writing_score'] > 90

# Gộp điều kiện bằng toán tử "|" (tương đương với OR trong lập trình)
df_it_nhat_1_mon_90 = df[cond_math | cond_reading | cond_writing]

# Tính toán số lượng và tỷ lệ %
so_luong = len(df_it_nhat_1_mon_90)
tong_so_hs = len(df)
ty_le_phan_tram = (so_luong / tong_so_hs) * 100

print(f"--- KẾT QUẢ THỐNG KÊ PHÂN KHÚC XUẤT SẮC (> 90) ---")
print(f"- Số lượng học sinh có ít nhất 1 môn > 90 điểm: {so_luong} học sinh.")
print(f"- Tỷ lệ chiếm: {ty_le_phan_tram:.2f}% trên tổng số {tong_so_hs} học sinh toàn trường.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Đọc file báo cáo dữ liệu của bạn
df = pd.read_csv('Bao_Cao_Du_Doan_Hoc_Sinh.csv')
df.columns = df.columns.str.strip().str.lower()

# ==========================================================
# THỰC HIỆN LOGIC PHÂN LOẠI CHI TIẾT 6 NHÓM (Chuẩn hệ số 0.265)
# ==========================================================
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['math_low'] = df['math_score'] < (0.265 * df['total_score'])
df['reading_low'] = df['reading_score'] < (0.265 * df['total_score'])
df['writing_low'] = df['writing_score'] < (0.265 * df['total_score'])

def phan_loai_chi_tiet_6_nhom(row):
    math, reading, writing = row['math_score'], row['reading_score'], row['writing_score']
    if math > 80 and reading > 80 and writing > 80: return 'Học tốt cả 3 môn'
    if math < 70 and reading < 70 and writing < 70: return 'Học kém cả 3 môn'
    if row['math_low']: return 'Học lệch - Yếu Toán'
    if row['reading_low']: return 'Học lệch - Yếu Đọc'
    if row['writing_low']: return 'Học lệch - Yếu Viết'
    return 'Học đều 3 môn'

df['Phân loại năng lực'] = df.apply(phan_loai_chi_tiet_6_nhom, axis=1)

# ==========================================================
# 2. KHỞI TẠO KHÔNG GIAN VẼ 3 BIỂU ĐỒ TRÒN SONG SONG
# ==========================================================
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Cấu hình danh sách môn và tiêu đề tương ứng
mon_hoc_list = ['math_score', 'reading_score', 'writing_score']
ten_mon_vn = ['Môn TOÁN (>90)', 'Môn ĐỌC (>90)', 'Môn VIẾT (>90)']

# Bảng màu cố định cho các nhóm năng lực để biểu đồ đồng bộ
colors_dict = {
    'Học tốt cả 3 môn': '#2ecc71',
    'Học đều 3 môn': '#3498db',
    'Học lệch - Yếu Toán': '#e67e22',
    'Học lệch - Yếu Đọc': '#f1c40f',
    'Học lệch - Yếu Viết': '#e74c3c',
    'Học kém cả 3 môn': '#95a5a6'
}

# Vòng lặp lọc dữ liệu và vẽ từng biểu đồ tròn
for i, mon in enumerate(mon_hoc_list):
    # Lọc học sinh có điểm môn hiện tại > 90
    df_sub = df[df[mon] > 90]
    data_pie = df_sub['Phân loại năng lực'].value_counts()
    
    # Lấy bảng màu tương ứng với các nhóm xuất hiện trong dữ liệu sub
    current_colors = [colors_dict[group] for group in data_pie.index]
    
    # Hiệu ứng tách nhẹ miếng bánh lớn nhất để tạo điểm nhấn
    explode = [0.05 if j == 0 else 0 for j in range(len(data_pie))]
    
    # Vẽ biểu đồ tròn trên trục (ax) tương ứng
    axes[i].pie(data_pie, 
                labels=data_pie.index, 
                autopct='%1.1f%%', 
                startangle=140, 
                colors=current_colors, 
                explode=explode,
                textprops={'fontsize': 10, 'fontweight': 'bold'})
    
    axes[i].set_title(f'Cấu trúc nhóm học sinh có\n{ten_mon_vn[i]}\n(Tổng: {len(df_sub)} HS)', 
                      fontsize=12, fontweight='bold', pad=10)

plt.suptitle('TỶ LỆ PHẦN TRĂM CÁC NHÓM NĂNG LỰC TRONG PHÂN KHÚC ĐIỂM ĐỈNH (>90)', 
             fontsize=15, fontweight='bold', y=1.05)
plt.tight_layout()
plt.show()
```

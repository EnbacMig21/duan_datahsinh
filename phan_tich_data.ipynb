### 1.1 Phân tính kết quả theo kết quả học tập.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu từ file của bạn
df = pd.read_csv('student_info_updated.csv')

# 2. Định nghĩa từ điển quy đổi (Mapping) chuẩn xác theo ảnh bạn gửi
mapping_edu = {
    1: 'High School',
    2: "Bachelor's",
    3: "Master's",
    4: 'PhD',
    '1': 'High School',
    '2': "Bachelor's",
    '3': "Master's",
    '4': 'PhD'
}

# Tạo cột nhãn chữ mới
df['parent_education_label'] = df['parent_education'].map(mapping_edu)
df['parent_education_label'] = df['parent_education_label'].fillna(df['parent_education'])

# 3. Tính điểm trung bình của từng môn theo nhãn chữ mới
subject_scores_by_parent = df.groupby('parent_education_label')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()
subject_scores_by_parent = subject_scores_by_parent.round(2)

# Sắp xếp thứ tự hiển thị từ thấp đến cao theo đúng logic học thuật
ordered_categories = ['High School', "Bachelor's", "Master's", 'PhD']
subject_scores_by_parent['parent_education_label'] = pd.Categorical(
    subject_scores_by_parent['parent_education_label'], 
    categories=ordered_categories, 
    ordered=True
)
subject_scores_by_parent = subject_scores_by_parent.sort_values('parent_education_label').reset_index(drop=True)

# Đổi tên các cột sang tiếng Việt để bảng in ra sạch sẽ
subject_scores_by_parent.columns = ['Cấp học phụ huynh', 'Điểm Toán TB', 'Điểm Đọc TB', 'Điểm Viết TB']

# 4. In kết quả số cụ thể ra màn hình
print("--- KẾT QUẢ SỐ CỤ THỂ: ĐIỂM 3 MÔN THEO CẤP HỌC PHỤ HUYNH ---")
print(subject_scores_by_parent.to_string(index=False))

# ==========================================================
# 5. TRỰC QUAN HÓA (Vẽ biểu đồ cột nhóm)
# ==========================================================
df_melted = pd.melt(subject_scores_by_parent, 
                    id_vars=['Cấp học phụ huynh'], 
                    value_vars=['Điểm Toán TB', 'Điểm Đọc TB', 'Điểm Viết TB'],
                    var_name='Môn học', 
                    value_name='Điểm số trung bình')

plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Vẽ biểu đồ cột nhóm
ax = sns.barplot(x='Cấp học phụ huynh', y='Điểm số trung bình', hue='Môn học', data=df_melted, palette='Set2')

# Cấu hình tiêu đề và nhãn trục
plt.title('THỐNG KÊ CHI TIẾT ĐIỂM 3 MÔN THEO CẤP HỌC PHỤ HUYNH', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Trình độ học vấn của phụ huynh', fontsize=11, labelpad=10)
plt.ylabel('Điểm trung bình (Thang điểm 100)', fontsize=11)
plt.ylim(0, 115) 

# Hiển thị số điểm cụ thể lên đầu từng cột
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'{p.get_height():.1f}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', 
                    xytext=(0, 8), 
                    textcoords='offset points', 
                    fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
mapping_lunch = {1: 'Standard', 0: 'Free or Reduced','1': 'Standard','0': 'Free or Reduced'}
df['lunch_type_label'] = df['lunch_type'].map(mapping_lunch)
# Nếu có giá trị nào không khớp, giữ nguyên giá trị gốc của cột lunch_type
df['lunch_type_label'] = df['lunch_type_label'].fillna(df['lunch_type'])
# 3. Tính điểm trung bình của từng môn theo nhãn mới tạo
scores_by_lunch = df.groupby('lunch_type_label')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()
scores_by_lunch = scores_by_lunch.round(2)
# Đổi tên các cột sang tiếng Việt để bảng hiển thị sạch sẽ
scores_by_lunch.columns = ['Chế độ ăn trưa', 'Điểm Toán TB', 'Điểm Đọc TB', 'Điểm Viết TB']
# 4. In kết quả số cụ thể ra màn hình để bạn đối chiếu
print("--- KẾT QUẢ SỐ CỤ THỂ: ĐIỂM 3 MÔN THEO CHẾ ĐỘ ĂN TRƯA ---")
print(scores_by_lunch.to_string(index=False))

df_melted = pd.melt(scores_by_lunch, 
                    id_vars=['Chế độ ăn trưa'], 
                    value_vars=['Điểm Toán TB', 'Điểm Đọc TB', 'Điểm Viết TB'],
                    var_name='Môn học', 
                    value_name='Điểm số trung bình')

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
ax = sns.barplot(x='Chế độ ăn trưa', y='Điểm số trung bình', hue='Môn học', data=df_melted, palette='Set1')
plt.title('SO SÁNH ĐIỂM SỐ TRUNG BÌNH THEO CHẾ ĐỘ ĂN TRƯA', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Chế độ ăn trưa của học sinh', fontsize=11)
plt.ylabel('Điểm trung bình (Thang điểm 100)', fontsize=11)
plt.ylim(0, 115) 
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'{p.get_height():.1f}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', 
                    xytext=(0, 8), 
                    textcoords='offset points', 
                    fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu từ file của bạn
df = pd.read_csv('student_info_updated.csv')

# 2. Gom nhóm theo giới tính và tính trung bình của Giờ học & Chuyên cần
# (Hãy đảm bảo tên các cột 'study_hours' và 'attendance_rate' trùng khớp với file của bạn)
gender_metrics = df.groupby('gender')[['study_hours', 'attendance_rate']].mean().reset_index()

# Làm tròn kết quả đến 2 chữ số thập phân
gender_metrics = gender_metrics.round(2)

# Đổi tên cột sang tiếng Việt để xuất bảng dữ liệu
gender_metrics_view = gender_metrics.copy()
gender_metrics_view.columns = ['Giới tính', 'Giờ tự học trung bình (tuần)', 'Tỷ lệ chuyên cần trung bình (%)']

# 3. In kết quả số cụ thể ra màn hình
print("--- KẾT QUẢ SỐ CỤ THỂ: GIỜ HỌC VÀ CHUYÊN CẦN THEO GIỚI TÍNH ---")
print(gender_metrics_view.to_string(index=False))

# ==========================================================
# 4. TRỰC QUAN HÓA (Vẽ 2 biểu đồ cạnh nhau để dễ so sánh)
# ==========================================================
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sns.set_style("whitegrid")

# --- Biểu đồ 1: Giới tính và Giờ tự học ---
ax1 = sns.barplot(x='gender', y='study_hours', data=gender_metrics, ax=axes[0], palette='Blues_d')
axes[0].set_title('GIỜ TỰ HỌC TRUNG BÌNH THEO GIỚI TÍNH', fontsize=12, fontweight='bold', pad=15)
axes[0].set_xlabel('Giới tính', fontsize=11)
axes[0].set_ylabel('Số giờ tự học / tuần', fontsize=11)

# Thêm số trên đầu cột cho biểu đồ 1
for p in ax1.patches:
    ax1.annotate(f'{p.get_height():.2f}h', 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha='center', va='center', xytext=(0, 8), textcoords='offset points', fontweight='bold')

# --- Biểu đồ 2: Giới tính và Tỷ lệ chuyên cần ---
ax2 = sns.barplot(x='gender', y='attendance_rate', data=gender_metrics, ax=axes[1], palette='Purples_d')
axes[1].set_title('TỶ LỆ CHUYÊN CẦN TRUNG BÌNH THEO GIỚI TÍNH', fontsize=12, fontweight='bold', pad=15)
axes[1].set_xlabel('Giới tính', fontsize=11)
axes[1].set_ylabel('Tỷ lệ chuyên cần (%)', fontsize=11)
axes[1].set_ylim(0, 110) # Tạo khoảng trống phía trên để hiển thị ký tự %

# Thêm số trên đầu cột cho biểu đồ 2
for p in ax2.patches:
    ax2.annotate(f'{p.get_height():.1f}%', 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha='center', va='center', xytext=(0, 8), textcoords='offset points', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu từ file của bạn
# Bạn hãy thay 'student_info_updated.csv' bằng tên file chính xác của bạn
df = pd.read_csv('student_info_updated.csv')

# 2. Tính điểm trung bình các môn Toán, Đọc, Viết theo từng khối lớp
# (Hãy đảm bảo tên các cột trùng khớp với file dữ liệu của bạn)
scores_by_grade = df.groupby('grade_level')[['math_score', 'reading_score', 'writing_score']].mean().reset_index()

# Làm tròn kết quả đến 2 chữ số thập phân để báo cáo đẹp hơn
scores_by_grade = scores_by_grade.round(2)

# Đổi tên các cột để bảng hiển thị rõ nghĩa tiếng Việt
scores_by_grade.columns = ['Khối lớp', 'Điểm Toán TB', 'Điểm Đọc TB', 'Điểm Viết TB']

# 3. In kết quả số cụ thể ra màn hình
print("--- KẾT QUẢ SỐ CỤ THỂ: ĐIỂM SỐ TRUNG BÌNH THEO KHỐI LỚP ---")
print(scores_by_grade.to_string(index=False))

# ==========================================================
# 4. TRỰC QUAN HÓA (Biến đổi dữ liệu và vẽ biểu đồ)
# ==========================================================
# Chuyển bảng dữ liệu sang dạng "Long format" để vẽ biểu đồ nhóm bằng Seaborn
df_melted = pd.melt(scores_by_grade, 
                    id_vars=['Khối lớp'], 
                    value_vars=['Điểm Toán TB', 'Điểm Đọc TB', 'Điểm Viết TB'],
                    var_name='Môn học', 
                    value_name='Điểm số trung bình')

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Vẽ biểu đồ cột nhóm
ax = sns.barplot(x='Khối lớp', y='Điểm số trung bình', hue='Môn học', data=df_melted, palette='Set2')

# Cấu hình tiêu đề và nhãn trục
plt.title('THỐNG KÊ ĐIỂM SỐ TRUNG BÌNH CÁC MÔN THEO KHỐI LỚP', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Khối lớp', fontsize=11)
plt.ylabel('Điểm trung bình (Thang điểm 100)', fontsize=11)
plt.ylim(0, 115) # Tạo khoảng trống phía trên để hiển thị số

# Hiển thị số điểm cụ thể lên đầu từng cột
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'{p.get_height():.2f}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', 
                    xytext=(0, 8), 
                    textcoords='offset points', 
                    fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 2. Đổi nhãn số thành chữ cho dễ hiểu trên biểu đồ
net_labels = {0: 'Không có Internet', 1: 'Có Internet'}
df['internet_label'] = df['internet_access'].map(net_labels)

# 3. Khởi tạo kích thước khung tranh
plt.figure(figsize=(8, 6))

# 4. Vẽ Biểu đồ hộp (Box Plot)
sns.boxplot(
    data=df, 
    x='internet_label', 
    y='study_hours', 
    hue='internet_label', 
    palette='viridis',  # Dùng tông màu hiện đại
    legend=False
)

# 5. Thiết lập tiêu đề và tên trục
plt.title('Phân bổ Số giờ tự học theo tình trạng Internet', fontsize=14, fontweight='bold')
plt.xlabel('Tình trạng truy cập Internet', fontsize=12)
plt.ylabel('Số giờ học (giờ/ngày)', fontsize=12)

plt.tight_layout()
plt.show()
```

Yếu tố truy cập internet tại nhà không ảnh hưởng quá lớn đến việc học tập của học sinh. Tuy nhiên Khi được truy cập internet số giờ tự học TB có tăng và số lượng học sinh đạt 4h học cũng cao hơn nhiều. 

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='internet_access', hue='final_result', palette='Set2')
plt.title('Tỷ lệ Pass/Fail theo Internet Access')
plt.show()
```

```python
activity_labels = {0: 'Không tham gia', 1: 'Có tham gia'}
df['extra_activities_label'] = df['extra_activities'].map(activity_labels)

# 3. Khởi tạo kích thước khung tranh cho 1 biểu đồ duy nhất
plt.figure(figsize=(8, 6))

# 4. Vẽ Biểu đồ hộp (Box Plot)
sns.boxplot(
    data=df, 
    x='extra_activities_label', 
    y='study_hours', 
    hue='extra_activities_label', 
    palette='Pastel1',
    legend=False
)

# 5. Thiết lập tiêu đề và tên các trục
plt.title('Phân bổ Số giờ tự học theo Hoạt động ngoại khóa', fontsize=14, fontweight='bold')
plt.xlabel('Hoạt động ngoại khóa', fontsize=12)
plt.ylabel('Số giờ học (giờ/ngày)', fontsize=12)

# 6. Tối ưu không gian và hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

#### Từ đây chúng ta có thể kết luận, các biến số không ảnh hưởng quá nhiều đến nhau hay còn có thể nói là khá độc lập tuyến tính. Tuy nhiên khi đi sâu vào phân tích về cột 'study_hours' thì thấy những học sinh tham gia hoạt động ngoại khóa, và được truy cập internet thì có xu hướng chăm học hơn, số giờ học TB nhiều hơn. Các yếu tố đó ảnh hưởng rất tích cực đến sự hứng thú học tập của người học.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu từ file của bạn
df = pd.read_csv('student_info_updated.csv')

# 2. Tạo cột mới tính Tổng điểm 3 môn của từng học sinh
# (Hãy đảm bảo tên các cột trùng khớp với file dữ liệu của bạn)
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']

# 3. Gom nhóm theo Khối lớp và tính trung bình cộng của Tổng điểm
total_score_by_grade = df.groupby('grade_level')['total_score'].mean().reset_index()

# Làm tròn kết quả đến 2 chữ số thập phân
total_score_by_grade = total_score_by_grade.round(2)

# Đổi tên cột cho rõ nghĩa tiếng Việt
total_score_by_grade.columns = ['Khối lớp', 'Điểm tổng trung bình (3 môn)']

# 4. In kết quả số cụ thể ra màn hình
print("--- KẾT QUẢ SỐ CỤ THỂ: ĐIỂM TỔNG TRUNG BÌNH 3 MÔN THEO KHỐI LỚP ---")
print(total_score_by_grade.to_string(index=False))

# ==========================================================
# 5. TRỰC QUAN HÓA (Vẽ biểu đồ cột đơn giản để đưa lên Slide)
# ==========================================================
plt.figure(figsize=(7, 5))
sns.set_style("whitegrid")

# Vẽ biểu đồ cột
ax = sns.barplot(x='Khối lớp', y='Điểm tổng trung bình (3 môn)', data=total_score_by_grade, palette='YlGnBu_r')

# Cấu hình tiêu đề và nhãn
plt.title('ĐIỂM TỔNG TRUNG BÌNH (TOÁN + ĐỌC + VIẾT) THEO KHỐI LỚP', fontsize=12, fontweight='bold', pad=15)
plt.xlabel('Khối lớp', fontsize=11)
plt.ylabel('Tổng điểm trung bình (Thang điểm 300)', fontsize=11)

# Thiết lập giới hạn trục Y (tối đa là 300 điểm)
plt.ylim(0, 330)

# Hiển thị số liệu cụ thể lên đầu mỗi cột
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 8), 
                textcoords='offset points', 
                fontsize=10, fontweight='bold', color='black')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu từ file của bạn
# Bạn hãy thay 'student_info_updated.csv' bằng tên file chính xác của bạn nếu cần
df = pd.read_csv('student_info_updated.csv')

# 2. Tính tỷ lệ chuyên cần trung bình theo từng khối lớp
# (Giả sử tên cột khối lớp là 'grade_level' và cột chuyên cần là 'attendance_rate')
attendance_by_grade = df.groupby('grade_level')['attendance_rate'].mean().reset_index()

# Đổi tên cột để bảng hiển thị rõ nghĩa hơn
attendance_by_grade.columns = ['Khối lớp', 'Tỷ lệ chuyên cần trung bình (%)']

# 3. In kết quả số cụ thể ra màn hình
print("--- KẾT QUẢ SỐ CỤ THỂ: TỶ LỆ CHUYÊN CẦN THEO KHỐI LỚP ---")
print(attendance_by_grade.to_string(index=False))

# ==========================================================
# 4. TRỰC QUAN HÓA (Vẽ biểu đồ để đưa lên Slide Gamma)
# ==========================================================
plt.figure(figsize=(8, 5))
sns.set_style("whitegrid")

# Vẽ biểu đồ cột hiển thị tỷ lệ chuyên cần
ax = sns.barplot(x='Khối lớp', y='Tỷ lệ chuyên cần trung bình (%)', data=attendance_by_grade, palette='Blues_d')

# Cấu hình tiêu đề và nhãn
plt.title('TỶ LỆ CHUYÊN CẦN TRUNG BÌNH THEO TỪNG KHỐI LỚP', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Khối lớp', fontsize=11)
plt.ylabel('Tỷ lệ chuyên cần (%)', fontsize=11)

# Giới hạn trục Y từ 0 đến 110 để tạo khoảng trống hiển thị số trên đầu cột
plt.ylim(0, 110) 

# Hiển thị số liệu phần trăm cụ thể lên đầu mỗi cột
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 8), 
                textcoords='offset points', 
                fontsize=10, fontweight='bold', color='black')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu từ file của bạn
# Bạn hãy thay 'student_info_updated.csv' bằng tên file chính xác của bạn
df = pd.read_csv('student_info_updated.csv')

# 2. Tính số giờ tự học trung bình theo từng khối lớp
# (Giả sử tên cột khối lớp là 'grade_level' và cột số giờ tự học là 'study_hours')
study_by_grade = df.groupby('grade_level')['study_hours'].mean().reset_index()

# Đổi tên cột để bảng hiển thị rõ nghĩa
study_by_grade.columns = ['Khối lớp', 'Số giờ tự học trung bình (Giờ)']

# 3. In kết quả số cụ thể ra màn hình
print("--- KẾT QUẢ SỐ CỤ THỂ: SỐ GIỜ TỰ HỌC THEO KHỐI LỚP ---")
print(study_by_grade.to_string(index=False))

# ==========================================================
# 4. TRỰC QUAN HÓA (Vẽ biểu đồ xu hướng đưa lên Slide Gamma)
# ==========================================================
plt.figure(figsize=(8, 5))
sns.set_style("whitegrid")

# Vẽ biểu đồ đường thể hiện xu hướng tự học qua các khối lớp
ax = sns.lineplot(x='Khối lớp', y='Số giờ tự học trung bình (Giờ)', data=study_by_grade, 
                  marker='o', markersize=8, color='crimson', linewidth=2.5)

# Cấu hình tiêu đề và nhãn
plt.title('XU HƯỚNG SỐ GIỜ TỰ HỌC TRUNG BÌNH THEO KHỐI LỚP', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Khối lớp', fontsize=11)
plt.ylabel('Số giờ tự học (Giờ)', fontsize=11)

# Giới hạn trục Y rộng hơn một chút so với giá trị lớn nhất để biểu đồ đẹp hơn
max_hours = study_by_grade['Số giờ tự học trung bình (Giờ)'].max()
plt.ylim(0, max_hours + 1.5) 

# Hiển thị số liệu cụ thể ngay tại các điểm nút (markers)
for x, y in zip(study_by_grade['Khối lớp'], study_by_grade['Số giờ tự học trung bình (Giờ)']):
    plt.text(x, y + 0.15, f'{y:.2f} giờ', ha='center', va='bottom', 
             fontsize=10, fontweight='bold', color='darkred')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu
df = pd.read_csv('student_info_updated.csv')

# 2. Khởi tạo khung tranh
plt.figure(figsize=(10, 6))

# 3. Vẽ biểu đồ cột
ax = sns.barplot(
    data=df, 
    x='grade_level', 
    y='study_hours', 
    hue='grade_level', 
    palette='Blues_d', 
    legend=False,
    order=[9, 10, 11, 12]
)

# 4. Thêm nhãn số liệu lên đầu mỗi cột để báo cáo chuyên nghiệp hơn
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=3)

# 5. Thiết lập tiêu đề và nhãn
plt.title('Số giờ tự học Trung bình theo Khối lớp', fontsize=14, fontweight='bold')
plt.xlabel('Khối lớp', fontsize=12)
plt.ylabel('Số giờ học trung bình (giờ/ngày )', fontsize=12)

# 6. Hiển thị
plt.tight_layout()
plt.show()
```

#### Chúng ta có thể thấy, sự chăm chỉ của học sinh khối 11 thấp nhất trong 4 khối. Các chỉ số về số giờ học TB, tỷ lệ chuyên cần và điểm số trung bình của khối 11 đều thấp nhất so với 3 khối còn lại. Chúng ta cần quan sát và tìm kiếm nguyên nhân của sự việc này.

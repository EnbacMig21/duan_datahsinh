## Làm sạch và chuẩn hóa dữ liệu

Đổi giá trị cột 'final_result':'Pass' thành 1, 'Fail' thành 0

```python
df['final_result'] = df['final_result'].map({'Pass': 1, 'Fail': 0})
df
```

Đổi giá trị cột internet_access: 'Yes' thành 1, 'No' thành 0

```python
df['internet_access'] = df['internet_access'].map({'Yes': 1, 'No': 0})
df
```

Đổi giá trị cột extra_activities: 'Yes' thành 1, 'No' thành 0

```python
df['extra_activities'] = df['extra_activities'].map({'Yes': 1, 'No': 0})
df
```

Đổi giá trị cột lunch_type	: 'Standard' thành 1, 'Free or reduced' thành 0

```python
df['lunch_type'] = df['lunch_type'].map({'Standard': 1, 'Free or reduced': 0})
df
```

Đổi giá trị cột parent_education : 'High School' thành 1, 'Bachelor's' thành 2, 'Master's' thành 3, 'PhD' thành 4

```python
df['parent_education'] = df['parent_education'].map({'High School': 1, "Bachelor's": 2, "Master's": 3, 'PhD': 4})
df
```

## 1. Phân tích các chỉ số chung 

Ma trận tương quan giữa các biến số.

```python
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Ma trận tương quan giữa các biến số')
plt.show()
```

Quan sát ma trận ta có thể thấy, các hệ số tương quan giữa các biến khá nhỏ chỉ dao động từ -0,07 đến 0,07. Từ đó có thể nhật xét rằng, các biến số độc lập với nhau mà không bị phụ thuộc. Chúng không gây ảnh hưởng quá nhiều cho nhau.

```python
df['average_score'] = df[['math_score', 'reading_score', 'writing_score']].mean(axis=1)

# 3. Khởi tạo khung tranh
plt.figure(figsize=(10, 6))

# 4. Vẽ biểu đồ phân tán kèm đường hồi quy
sns.regplot(
    data=df, 
    x='study_hours', 
    y='average_score', 
    scatter_kws={'alpha':0.5, 'color': '#2a9d8f'}, # Độ mờ của các điểm
    line_kws={'color': '#e76f51'}                 # Màu của đường xu hướng
)

# 5. Thiết lập tiêu đề và nhãn
plt.title('Mối liên hệ giữa Số giờ tự học và Điểm trung bình', fontsize=14, fontweight='bold')
plt.xlabel('Số giờ tự học (giờ/ngày)', fontsize=12)
plt.ylabel('Điểm trung bình (Toán, Đọc, Viết)', fontsize=12)

# 6. Hiển thị
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Từ ma trận tương quan phía trên, kèm theo biểu đồ ta có thể đưa ra khẳng định như số giờ học TB không liên quan đến điểm số của học sinh.

#### Tương tự như phần 2, chúng ta có thể dễ dàng nhận ra rằng học sinh có bố mẹ là Master's học tập không hiệu quả. Các chỉ số về số giờ học TB, tỷ lệ chuyên cần và điểm số trung bình của con cái Master's thấp hơn so với các cấp độ còn lại. Mặc dù số lượng của học sinh theo từng yếu tố đều tương tự như nhau nên càng phản ánh rõ nét. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu
df = pd.read_csv('student_info_updated.csv')

# 2. Chuyển đổi cấu trúc dữ liệu (Melt) để vẽ biểu đồ nhóm
df_melted = pd.melt(
    df, 
    id_vars=['grade_level'], 
    value_vars=['math_score', 'reading_score', 'writing_score'],
    var_name='Mon_Hoc', 
    value_name='Diem_So'
)

# Đổi tên nhãn hiển thị sang tiếng Việt cho chuyên nghiệp
ten_mon_hoc = {
    'math_score': 'Toán',
    'reading_score': 'Đọc',
    'writing_score': 'Viết'
}
df_melted['Mon_Hoc'] = df_melted['Mon_Hoc'].map(ten_mon_hoc)

# 3. Khởi tạo khung tranh
plt.figure(figsize=(12, 7))

# 4. Vẽ biểu đồ cột nhóm (hue='Mon_Hoc')
ax = sns.barplot(
    data=df_melted, 
    x='grade_level', 
    y='Diem_So', 
    hue='Mon_Hoc', 
    palette='Set2', # Bộ màu pastel phân biệt rất tốt giữa 3 môn
    errorbar=None,
    order=[9, 10, 11, 12]
)

# 5. Thêm nhãn điểm số chính xác lên đầu mỗi cột
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=3)

# 6. Thiết lập tiêu đề và nhãn trục
plt.title('Điểm số Trung bình các môn Toán - Đọc - Viết theo từng Khối lớp', fontsize=14, fontweight='bold')
plt.xlabel('Khối lớp', fontsize=12)
plt.ylabel('Điểm số trung bình', fontsize=12)
plt.ylim(0, 110) # Tạo khoảng trống phía trên để không bị đè chữ
plt.legend(title='Môn học', loc='upper right')

# 7. Hiển thị
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu
df = pd.read_csv('student_info_updated.csv')

# 2. Tạo một cột mới chia điểm Toán thành các khoảng (mỗi khoảng 5 điểm)
# Ví dụ: Học sinh được 72 hay 74 điểm đều gộp chung vào nhóm 70-75
df['math_score_bin'] = (df['math_score'] // 5) * 5

# 3. Khởi tạo khung tranh
plt.figure(figsize=(12, 6))

# 4. Vẽ biểu đồ đường theo khoảng điểm
sns.lineplot(
    data=df, 
    x='math_score_bin', 
    y='attendance_rate', 
    color='#d93025', # Màu đỏ nổi bật
    linewidth=2.5,
    marker='o', # Thêm các chấm tròn tại mỗi khoảng điểm cho dễ nhìn
    errorbar=None
)

# 5. Thiết lập tiêu đề và nhãn
plt.title('Xu hướng Tỷ lệ Chuyên cần theo các Khoảng Điểm Toán', fontsize=14, fontweight='bold')
plt.xlabel('Khoảng Điểm số môn Toán (Trục X)', fontsize=12)
plt.ylabel('Tỷ lệ chuyên cần trung bình (%) (Trục Y)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# 6. Hiển thị
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu
df = pd.read_csv('student_info_updated.csv')

# 2. Tính tỷ lệ chuyên cần trung bình của TOÀN BỘ học sinh trong trường
attendance_mean_all = df['attendance_rate'].mean()
print(f"Tỷ lệ chuyên cần trung bình toàn trường: {attendance_mean_all:.2f}%")

# 3. Khởi tạo khung tranh
plt.figure(figsize=(12, 6))

# 4. Vẽ biểu đồ đường xu hướng (Gom nhóm tính trung bình theo từng điểm Toán để đường mịn hơn)
sns.lineplot(
    data=df, 
    x='math_score', 
    y='attendance_rate', 
    color='#1a73e8', 
    linewidth=2,
    label='Tỷ lệ chuyên cần theo điểm Toán',
    errorbar=None
)

# 5. Vẽ ĐƯỜNG TRUNG BÌNH TOÀN TRƯỜNG (Đường ngang nét đứt)
plt.axhline(
    y=attendance_mean_all, 
    color='#d93025',      # Màu đỏ nổi bật
    linestyle='--',       # Nét đứt (dashed line)
    linewidth=1, 
    label=f'Trung bình toàn trường ({attendance_mean_all:.2f}%)'
)

# 6. Thiết lập tiêu đề, nhãn và chú thích
plt.title('Mối liên hệ giữa Điểm Toán và Tỷ lệ Chuyên cần', fontsize=14, fontweight='bold')
plt.xlabel('Điểm số môn Toán (Trục X)', fontsize=12)
plt.ylabel('Tỷ lệ chuyên cần (%) (Trục Y)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

# Hiển thị bảng chú thích (Legend) để phân biệt 2 đường
plt.legend(loc='lower right', fontsize=11)

# 7. Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu
df = pd.read_csv('student_info_updated.csv')

# 2. Tính tỷ lệ chuyên cần trung bình của TOÀN BỘ học sinh (dùng chung cho cả 3 biểu đồ)
attendance_mean_all = df['attendance_rate'].mean()

# 3. Khởi tạo khung tranh gồm 1 hàng và 3 cột, dùng chung trục Y (sharey=True) để dễ so sánh
fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=True)

# Danh sách các môn học, màu sắc và vị trí cột tương ứng
mon_hoc = [
    {'col': 'math_score', 'name': 'Môn Toán', 'color': '#1a73e8'},
    {'col': 'reading_score', 'name': 'Môn Đọc', 'color': '#ff7f0e'},
    {'col': 'writing_score', 'name': 'Môn Viết', 'color': '#2ca02c'}
]

# 4. Vòng lặp tự động vẽ 3 biểu đồ
for i, mon in enumerate(mon_hoc):
    # Vẽ biểu đồ đường xu hướng của từng môn
    sns.lineplot(
        data=df, 
        x=mon['col'], 
        y='attendance_rate', 
        color=mon['color'], 
        linewidth=2,
        ax=axes[i],
        errorbar=None,
        label=f"Xu hướng {mon['name']}"
    )
    
    # Vẽ đường trung bình toàn trường (đường ngang nét đứt màu đỏ)
    axes[i].axhline(
        y=attendance_mean_all, 
        color='#d93025', 
        linestyle='--', 
        linewidth=1.8, 
        label=f'TB toàn trường ({attendance_mean_all:.2f}%)'
    )
    
    # Cấu hình chi tiết cho từng biểu đồ nhỏ
    axes[i].set_title(f"Mối liên hệ với {mon['name']}", fontsize=13, fontweight='bold')
    axes[i].set_xlabel(f"Điểm số {mon['name']}")
    axes[i].grid(True, linestyle='--', alpha=0.5)
    axes[i].legend(loc='lower right', fontsize=10)

# Cấu hình trục Y cho biểu đồ đầu tiên (các biểu đồ sau sẽ tự động ăn theo)
axes[0].set_ylabel('Tỷ lệ chuyên cần (%)', fontsize=12)

# Tiêu đề tổng của cả bức tranh
plt.suptitle('Xu hướng Tỷ lệ Chuyên cần theo Điểm số của từng Môn học', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu
df = pd.read_csv('student_info_updated.csv')

# 2. Tính số giờ tự học trung bình của TOÀN BỘ học sinh trong trường (đường baseline)
study_hours_mean_all = df['study_hours'].mean()
print(f"Số giờ tự học trung bình toàn trường: {study_hours_mean_all:.2f} giờ/tuần")

# 3. Khởi tạo khung tranh gồm 1 hàng và 3 cột, dùng chung trục Y (sharey=True)
fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=True)

# Danh sách cấu hình cho 3 môn học
mon_hoc = [
    {'col': 'math_score', 'name': 'Môn Toán', 'color': '#1a73e8'},
    {'col': 'reading_score', 'name': 'Môn Đọc', 'color': '#ff7f0e'},
    {'col': 'writing_score', 'name': 'Môn Viết', 'color': '#2ca02c'}
]

# 4. Vòng lặp tự động vẽ 3 biểu đồ đường
for i, mon in enumerate(mon_hoc):
    # Vẽ biểu đồ đường xu hướng (Gom nhóm tính số giờ học trung bình theo từng mức điểm)
    sns.lineplot(
        data=df, 
        x=mon['col'], 
        y='study_hours', 
        color=mon['color'], 
        linewidth=2,
        ax=axes[i],
        errorbar=None,
        label=f"Xu hướng {mon['name']}"
    )
    
    # Vẽ đường số giờ học trung bình toàn trường (đường ngang nét đứt màu tím/đỏ)
    axes[i].axhline(
        y=study_hours_mean_all, 
        color='#7b1fa2', # Màu tím đậm để phân biệt với biểu đồ chuyên cần trước đó
        linestyle='--', 
        linewidth=1.8, 
        label=f'TB toàn trường ({study_hours_mean_all:.2f} giờ)'
    )
    
    # Cấu hình chi tiết cho từng ô đồ thị
    axes[i].set_title(f"Mối liên hệ với {mon['name']}", fontsize=13, fontweight='bold')
    axes[i].set_xlabel(f"Điểm số {mon['name']}")
    axes[i].grid(True, linestyle='--', alpha=0.5)
    axes[i].legend(loc='upper left', fontsize=10)

# Cấu hình trục Y cho biểu đồ đầu tiên
axes[0].set_ylabel('Số giờ tự học trung bình (giờ/tuần)', fontsize=12)

# Tiêu đề tổng của cả bức tranh
plt.suptitle('Xu hướng Số giờ Tự học theo Điểm số của từng Môn học', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 1. Đọc dữ liệu
df = pd.read_csv('student_info_updated.csv')

# Kiểm tra các giá trị chữ đang có trong cột giới tính (ví dụ tên cột là 'gender')
print("Các giá trị gốc:", df['gender'].unique())

# 2. Tạo bản đồ chuyển đổi (Mapping)
# Bạn hãy đổi chữ 'Male', 'Female' thành đúng chữ có trong file của bạn (ví dụ: 'Nam', 'Nữ')
gioi_tinh_map = {
    'Female': 0,
    'Male': 1,
    'Other': 2
}

# 3. Ép kiểu chuyển đổi và tạo cột mới (hoặc ghi đè cột cũ)
df['gender_encoded'] = df['gender'].map(gioi_tinh_map)

# 4. Lưu lại vào file CSV
df.to_csv('student_info_updated.csv', index=False)
```
